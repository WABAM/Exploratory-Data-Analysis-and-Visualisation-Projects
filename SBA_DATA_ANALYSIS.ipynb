## Outliers
## The five-number summary

```python
sns.boxplot(x = df['CurrentApprovalAmount'])
```

```python
##The five-number summary divides the data into sections that each contain approximately 25 percent of the data in that set.
#The five-number summary is the minimum, first quartile, median, third quartile, and maximum

# Min = Minimum value
# Q1 = The first quartile is the median of the data points to the left of the median
# Median = is the value separating the higher half from the lower half of a data sample, a population or distribution
# Q3 = The third quartile is the median of the data points to the right of the median
# Max = Max value

#About 75 percent of the CurrentApprovalAmount is greater than 8700 dollars per the boxplot above.
```

```python
plt.figure(figsize=(15,6)) # this creates a figure 8 inch wide, 4 inch high
sns.boxplot(x="LoanStatus", y="CurrentApprovalAmount", data=df)
```

```python
sns.distplot(df['CurrentApprovalAmount']/10**3)
```

```python
#Deviates from the normal distribution.
#Have a positive skewness
#skewness and kurtosis
#Basically getting the body measure of the above diagram
```

```python
#df['CurrentApprovalAmount'].skew()
#df['CurrentApprovalAmount'].kurt()

###Modified 
print("Skewness: %f" % df['CurrentApprovalAmount'].skew())
print("Kurtosis: %f" % df['CurrentApprovalAmount'].kurt())
```
